# Matrices: lista de listas

Una matriz es simplemente una lista donde cada elemento es otra lista. 

Si una lista es una estructura de una dimensión, una matriz es una estructura de dos dimensiones

# Matrices estáticas

```python
# Declarar una matriz de 2 x 3
matriz = [[1, 3, 2], [3, 5, 1]]
```

```python
# Mostar la matriz en la consola
print(f"La matriz es: {matriz}")
```

```python
# Mostar la matriz en la consola mediante un bucle
print(f"La matriz es mediante un bucle:")
for fila in matriz:
    print(fila)
```

```python
# Para obtener el número de filas de una matriz hacemos
n_filas = len(matriz)
print(f"El número de filas de la matriz es: {n_filas}")
```

```python
# Para obtener el número de columnas de una matriz hacemos
n_columnas = len(matriz[0]) 
print(f"El número de columnas de la matriz es: {n_columnas}")
```

```python
# Mostrar sólo una fila de la matriz
print(f"La primera fila de la matriz es: {matriz[0]}")
```

```python
# Mostrar solo un elemento de la matriz
print(f"El elemento en la posición (0, 1) es: {matriz[0][1]}")
```

# Mostrar matrices

```python
# Creamos una función para mostrar la matriz de forma más legible
def mostrar_matriz(matriz):
    for fila in matriz:
        print(fila)
```

# Matrices especiales

Existen tres matrices especiales que se caracterízan los sus valores:
* matriz nula: aquella que tiene todos sus valores en 0
* matriz de unos: aquella que tiene todos sus valores en 1
* matriz identidad: aquella cuyos valores en la diagonal valen 1 y el resto 0

## Matriz nula

Vamos a desarrollar una función que genere una matriz nula de n x m dimensiones

```python
def matriz_nula(n_filas, n_columnas):
    matriz = []
    for _ in range(n_filas):
        fila = []
        for _ in range(n_columnas):
            fila.append(0)
        matriz.append(fila)
    return matriz
```

```python
# Crear una matriz nula de 3 x 4
matriz1 = matriz_nula(3, 4)
mostrar_matriz(matriz1)
```

## Matriz de unos

```python
def matriz_unos(n_filas, n_columnas):
    matriz = []
    for _ in range(n_filas):
        fila = []
        for _ in range(n_columnas):
            fila.append(1)
        matriz.append(fila)
    return matriz
```

```python
# Crear una matriz de unos de 3 x 4
matriz2 = matriz_unos(3, 4)
mostrar_matriz(matriz2)
```

## Matriz identidad

```python
def matriz_identidad(n_filas, n_columnas):
    matriz = []
    for i in range(n_filas):
        fila = []
        for j in range(n_columnas):
            fila.append(1 if i == j else 0)
        matriz.append(fila)
    return matriz
```

```python
# Crear una matriz identidad de 3 x 4
matriz3 = matriz_identidad(3, 4)
mostrar_matriz(matriz3)
```

## Matriz random

```python
import random

def matriz_random(n_filas, n_columnas):
    matriz = []
    for _ in range(n_filas):
        fila = []
        for _ in range(n_columnas):
            fila.append(random.randint(0, 9))  # Agrega un número aleatorio entre 0 y 9
        matriz.append(fila)
    return matriz
```

```python
# Crear una matriz random de 3 x 4
matriz4 = matriz_random(3, 4)
mostrar_matriz(matriz4)
```

# Operaciones con matrices

## Suma de matrices

La suma de matrices esta permitida siempre y cuando tenga la misma dimensión.

```python
# Crear una función para sumar dos matrices
def sumar_matrices(matriz_a, matriz_b):
    if len(matriz_a) != len(matriz_b) or len(matriz_a[0]) != len(matriz_b[0]):
        print("Imposible sumar matrices de diferente dimensión")
        return
    n_filas = len(matriz_a)
    n_columnas = len(matriz_a[0])
    suma = matriz_nula(n_filas, n_columnas)
    for i in range(len(matriz_a)):
        for j in range(len(matriz_a[0])):
            suma[i][j] = matriz_a[i][j] + matriz_b[i][j]
    return suma
```

```python
# Evaluar la función de suma de matrices
matriz_a = matriz_random(2, 2)
matriz_b = matriz_random(2, 2)
suma = sumar_matrices(matriz_a, matriz_b)
mostrar_matriz(suma)
```

## Multiplicación por un escalar

La multiplicación de una matriz por un escalar no requiere ninguna validación

```python
def matriz_por_escalar(matriz, escalar):
    n_filas = len(matriz)
    n_columnas = len(matriz[0])
    resultado = matriz_nula(n_filas, n_columnas)
    for i in range(len(matriz)):
        for j in range(len(matriz[0])):
            resultado[i][j] = matriz[i][j] * escalar
    return resultado
```

```python
# Evaluar la multiplicación de una matriz por un escalar
matriz_a = matriz_random(2, 2)
escalar = 3
resultado = matriz_por_escalar(matriz_a, escalar)
mostrar_matriz(resultado)
```

## Multiplicación entre matrices

Dadas dos matrices, matriz_a y matriz_b, el producto matricial entre ellas está definido únicamente si el número de columnas de la matriz_a es igual al número de filas de la matriz_b, es decir si y solo si, matriz_a (n x m) y matriz_b (m x p)

Como resultado del producto de matriz_a (n x m) y la matriz_b (m x p), se obtiene la matriz_c (n x p)

```python
def producto_matrices(matriz_a, matriz_b):
    # Obtener las dimensiones de las matrices
    n = len(matriz_a) # numero de filas de la matriz_a
    m = len(matriz_a[0]) # numero de columnas de la matriz_a
    o = len(matriz_b) # numero de filas de la matriz_b
    p = len(matriz_b[0]) # numero de columnas de la matriz_b

    # Validar que sea posible el producto entre las matrices
    if m != o:
        print("Imposible multiplicar matrices con dimensiones incompatibles")
        return
    
    # cerar una matriz de ceros para almacenar el resultado
    producto = matriz_nula(n, p)

    # i se desplaza por las filas de la matriz_c
    for i in range(n):
        # k se desplaza por las columnas de la matriz_c
        for k in range(p):
            # j se desplaza por las columnas de la matriz_a y las filas de la matriz_b
            for j in range(m):
                producto[i][k] += matriz_a[i][j] * matriz_b[j][k]
    return producto 
```

```python
# Evaluar la multiplicación de una matriz por un escalar
matriz_a = matriz_random(2, 3)
mostrar_matriz(matriz_a)
matriz_b = matriz_random(3, 2)
mostrar_matriz(matriz_b)
resultado = producto_matrices(matriz_a, matriz_b)
mostrar_matriz(resultado)
```

## Crear matriz normal

```python
import numpy as np
import matplotlib.pyplot as plt
from math import exp, pi, sqrt


def matriz_normal(nFilas, nColumnas=None):
    if nColumnas is None:
        nColumnas = nFilas
    matriz = [[0] * nColumnas for _ in range(nFilas)]

    # Parámetros de la normal
    mu_x, mu_y = 0, 0
    sigma_x, sigma_y = 1, 1

    # Rango de coordenadas
    x_vals = np.linspace(-3, 3, nColumnas)
    y_vals = np.linspace(-3, 3, nFilas)

    for i in range(nFilas):
        for j in range(nColumnas):
            x = x_vals[j]
            y = y_vals[i]
            # Fórmula de la normal bivariada
            z = (1 / (2 * pi * sigma_x * sigma_y)) * exp(
                -(((x - mu_x) ** 2) / (2 * sigma_x**2) + ((y - mu_y) ** 2) / (2 * sigma_y**2))
            )
            matriz[i][j] = z
    return matriz
```

## Crear plot

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt


def plot_matriz(matriz):
    Z = np.array(matriz)  # conversión para graficar
    n, m = Z.shape
    x_vals = np.linspace(-3, 3, m)
    y_vals = np.linspace(-3, 3, n)
    X, Y = np.meshgrid(x_vals, y_vals)
    fig = plt.figure(figsize=(7, 5))
    ax = fig.add_subplot(111, projection="3d")
    ax.plot_surface(X, Y, Z, edgecolor="k")  # simple y claro
    ax.set_title("Plot")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Altura")
    plt.show()
```
